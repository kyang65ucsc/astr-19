```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.poisson(1000, 1000)
plt.hist(x, bins=100, edgecolor='orange', color='cyan')
plt.xlabel('x')
plt.ylabel('N per bin')
plt.title('Poisson Distribution')

plt.savefig('session10')

plt.show()
```
